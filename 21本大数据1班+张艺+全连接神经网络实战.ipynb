```python
import torch
import torch.nn as nn

class MyNet(nn.Module):
    def __init__(self):
        super(MyNet, self).__init__()
        self.fc1 = nn.Linear(10, 64)  # 假设输入维度是10，第一层隐藏层有64个神经元
        self.fc2 = nn.Linear(64, 32)  # 第二层隐藏层有32个神经元
        self.fc3 = nn.Linear(32, 16)  # 第三层隐藏层有16个神经元
        self.fc4 = nn.Linear(16, 5)   # 输出层有5个神经元，对应5分类

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.relu(self.fc3(x))
        x = self.fc4(x)
        return x
```

```python
net = MyNet()
input_data = torch.randn(1, 10)  # 单个样本，维度为10
output = net(input_data)
print(output)
```

```python
criterion = nn.CrossEntropyLoss()
target = torch.randint(0, 5, (1,))  # 随机生成一个0到4之间的目标类别
loss = criterion(output, target)
loss.backward()
for name, param in net.named_parameters():
    if param.grad is not None:
        print(name, param.grad)
```
